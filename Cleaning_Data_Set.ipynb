```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import missingno as msno
import datetime as dt
```

```python
Data = pd.read_csv('/content/ML_Class_Demo.csv', delimiter = ",")
Data.columns = ((Data.columns.str).replace("^ ","")).str.replace(" $","")
```

```python
#replacing values in a data set with a new value
Data = Data.replace('?', np.NaN)
```

```python
#filling null values
Data.fillna(0) #fill nulls with a value
Data.fillna(method='ffill') #fill based on a method chosen from  {‘backfill’, ‘bfill’, ‘pad’, ‘ffill’, None}
Data.fillna(value={'A': 0, 'B': 1, 'C': 2, 'D': 3}) #fill each feature with a different value
```

```python
Data.duplicated()
```

```python
Data.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
#dropping null values
Data.dropna() #drop rows
Data.dropna(axis='columns') #drop columns
Data.dropna(axis='rows', thresh=3) # drop rows with fewer non-nulls than the threshhold
```
